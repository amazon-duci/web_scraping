# Creating a QUERY string

## GET Request
##### Retrieves data from the server with the URL parameters

```python
url_get = "http://httpbin.org/get" # Assigning the base URL and the route ('/get')
payload = {"name" : "Joseph", "ID": "123"} # creating parameters in a dictionary structure

# Assigning the imported request (first line) and using the get function
req = requests.get(url_get, params = payload)
```

```python
# to printout the url
req.url
```

```python
print(req.request.body)
```

```python
req.text
```

```python
req.headers["Content-Type"]
```

```python
# to format it using json()
req.json()
```

```python
# You use the key "ARGS" to get the payload
req.json()['args']
```

## Post Request
##### USed to send data to a server. But sends data in a body not the url

```python
url_post = "http://httpbin.org/post"
r_post = requests.post(url_post, params = payload)
```

```python
r_post.url
```

```python
# The post request has a body
arrb = r_post.request.body
arrb
```

```python
# you use the key "form" to get the payload
r_post.json()['form']
```

```python
# Importing Beautiful Soup
from bs4 import BeautifulSoup as bs
```

```python
# fetching and parsing HTML

# specify the URL
url = 'https://en.wikipedia.org/wiki/IBM'

# Send the GET request
r_get = requests.get(url)
r_get
```

```python
# Store the HTML content in a variable
html_content = r_get.text
```

```python
# Create Beautiful soup Object to parse the HTML
soup = bs(html_content, 'html.parser')
```

```python
# Display a snipet of the HTML content
print(html_content[:500])
```

### Navigating the HTML structure
#### You can use methods like <b>'find_all'</b> to filter and extract specific HTML elements

```python
# To find all anchor tags <a>
a_link = soup.find_all('a')
# a_link

table_row = soup.find_all(name='tr')

# to loop through and print all
# for link in a_link:
    # print(link)
```

```python
# Loading the necessary libraries
from randomuser import RandomUser
import pandas as pd
```

```python
# Creating a random user object
r_user = RandomUser()
```

```python
# generating a list of random users
my_list = r_user.generate_users(10)
my_list
```

```python
# to get a full name, we call the get_full_name() function
name = r_user.get_full_name()
full_name = {}
for user in my_list:
    print(f"{user.get_full_name()} {user.get_email()}")
```

```python
def get_users(x):
    users = []
    for user in RandomUser.generate_users(x):
        users.append({'Name': user.get_full_name(), 'Gender': user.get_gender(), 'City': user.get_city(), 'State': user.get_state(), 'Email': user.get_email(), 'DOB': user.get_dob(), 'Picture': user.get_picture()})
    return pd.DataFrame(users)
```

```python
get_users(10)
```

# Fruityvice API

```python
import json
```

```python
# obtaining Fruityvice API data using requests.get("URL") function in a JSON format
data = requests.get("https://web.archive.org/web/20240929211114/https://fruityvice.com/api/fruit/all")
data.headers['Content-Type']
```

```python
# retrieving result using json.loads()
results = json.loads(data.text)
```

```python
# converting json data into pandas DataFrame
pd.DataFrame(results)
```

```python
# The result is in a nested json format ===> Nutrition column
# To flatten it
df2 = pd.json_normalize(results)
```

```python
df2
```

```python
cherry = df2.loc[df2['name'] == 'Cherry']
(cherry.iloc[0]['family']), (cherry.iloc[0]['genus'])
```

```python
# Retrieving the calories of banana
banana = df2.loc[df2['name'] == 'Banana']
(banana.iloc[0]['nutritions.calories'])
```
